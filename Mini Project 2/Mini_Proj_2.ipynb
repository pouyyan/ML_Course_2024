### Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(W, x):
  return 1/(1 + np.exp(-(W*x)))
x = np.arange(-10,10,0.1)
y_sigmoid = sigmoid(1,x)

plt.plot(x, y_sigmoid)
plt.title('Sigmoid')
plt.xlabel('Input(x)')
plt.grid(True, linestyle = '--')
plt.show()
```

### ReLU

```python
def ReLU(x):
  return np.maximum(0, x)
x = np.arange(-10,10,0.1)
y = ReLU(x)
plt.plot(x, y)
plt.title('ReLU')
plt.xlabel('Input(x)')
plt.grid(True, linestyle = '--')
plt.show()
```

### Derivative of Sigmoid and ReLU

```python
# Define derivative functions
def sigmoid_derivative(x):
  return sigmoid(1, x)*(1 - sigmoid(1, x))

def ReLU_derivative(x):
  return np.where(x > 0, 1.0, 0.0)  # Using NumPy's where function

x = np.arange(-10,10,0.1)
y_d_sigmoid = sigmoid_derivative(x)
y_d_ReLU = ReLU_derivative(x)

plt.figure()
plt.plot(x, y_d_ReLU, label='ReLU Derivative', linestyle = '--')
plt.plot(x, y_d_sigmoid, label='Sigmoid Derivative', color = 'cyan')

plt.xlabel("X")
plt.ylabel("Derivative")
plt.title("Derivatives of Sigmoid and ReLU Functions")
plt.legend()
plt.grid(True, linestyle = '--')
plt.show()
```

### ELU

```python
def ELU(x, alpha = 1):
  return np.where(x >= 0, x, alpha*(np.exp(x)-1))

def D_ELU(x, alpha = 1):
  return np.where(x>=0, 1, alpha*np.exp(x))

x = np.arange(-10,10,0.1)
y_ELU = ELU(x)
y_D_ELU = D_ELU(x)

plt.figure()
plt.plot(x, y_ELU, label='ELU', linestyle = '--')
plt.plot(x, y_D_ELU, label='ELU Derivative', color = 'cyan')
plt.xlabel("X")
plt.ylabel("Derivative")
plt.title("Derivatives of Sigmoid and ReLU Functions")
plt.legend()
plt.grid(True, linestyle = '--')
plt.show()
```

## Question 1 Part 3

```python
#define muculloch pitts
class McCulloch_Pitts_neuron():
  def __init__(self , weights , threshold):
    self.weights = np.array(weights)
    self.threshold = np.array(threshold)

  def model(self , x):
    if self.weights @ x >= self.threshold:
        return 1
    else:
        return 0

#define model for dataset
def Area(x, y):
  N_11 = McCulloch_Pitts_neuron([2, -1], 2)
  N_12 = McCulloch_Pitts_neuron([-2,-1], -6)
  N_13 = McCulloch_Pitts_neuron([0, 1], 0)

  N_21 = McCulloch_Pitts_neuron([1, 1, 1], 3)

  z11 = N_11.model(np.array([x, y]))
  z12 = N_12.model(np.array([x, y]))
  z13 = N_13.model(np.array([x, y]))

  z21 = N_21.model(np.array([z11, z12, z13]))

  # 3 bit output
  # return str(z1) + str(z2)
  return list([z21])
```

### M*N Window

```python
# M is the number of samples
M = 200
# N is the window size
N = 400
Output = []
# Normal = 0, Fault_IR = 1, Fault_B = 2, Fault_OR = 3
np.random.seed(0)

cnt1 = 0
index_list = []
while cnt1 < M:
  index = np.random.choice(len(Normal_data_df))
  if len(Normal_data_df) - index > N:
    cnt1 += 1
    index_list.append(index)
    Output.append(0)

cnt2 = 0
index_list1 = []
while cnt2 < M:
  index1 = np.random.choice(len(Fault_data_IR_df))
  if len(Fault_data_IR_df) - index1 > N:
    cnt2 += 1
    index_list1.append(index1)
    Output.append(1)

cnt3 = 0
index_list2 = []
while cnt3 < M:
  index2 = np.random.choice(len(Fault_data_B_df))
  if len(Fault_data_B_df) - index2 > N:
    cnt3 += 1
    index_list2.append(index2)
    Output.append(2)

cnt4 = 0
index_list3 = []
while cnt4 < M:
  index3 = np.random.choice(len(Fault_data_OR_df))
  if len(Fault_data_OR_df) - index3 > N:
    cnt4 += 1
    index_list3.append(index3)
    Output.append(3)

Normal_data_numpy = Normal_data_df.to_numpy()
Normal_data_numpy = Normal_data_numpy.reshape(-1,len(Normal_data_numpy))

Fault_data_IR_df_numpy = Fault_data_IR_df.to_numpy()
Fault_data_IR_df_numpy = Fault_data_IR_df_numpy.reshape(-1,len(Fault_data_IR_df_numpy))

Fault_data_B_df_numpy = Fault_data_B_df.to_numpy()
Fault_data_B_df_numpy = Fault_data_B_df_numpy.reshape(-1,len(Fault_data_B_df_numpy))

Fault_data_OR_df_numpy = Fault_data_OR_df.to_numpy()
Fault_data_OR_df_numpy = Fault_data_OR_df_numpy.reshape(-1,len(Fault_data_OR_df_numpy))
```

```python
Array_Normal_data = np.zeros((M,N))
for index, element in enumerate(index_list):
  seq_Normal_data = Normal_data_numpy[0, element:element+N]
  Array_Normal_data[index-1,:] = seq_Normal_data
  # Labeling Fault output as 0

Array_Fault_data_IR = np.zeros((M,N))
for index1, element1 in enumerate(index_list1):
  seq_Fault_data_IR = Fault_data_IR_df_numpy[0, element1:element1+N]
  Array_Fault_data_IR[index1-1,:] = seq_Fault_data_IR
  # Labeling Fault Inner Race output as 1

Array_Fault_data_B = np.zeros((M,N))
for index2, element2 in enumerate(index_list2):
  seq_Fault_data_B = Fault_data_B_df_numpy[0, element2:element2+N]
  Array_Fault_data_B[index2-1,:] = seq_Fault_data_B
  # Labeling Fault Ball output as 2

Array_Fault_data_OR = np.zeros((M,N))
for index3, element3 in enumerate(index_list3):
  seq_Fault_data_OR = Fault_data_OR_df_numpy[0, element3:element3+N]
  Array_Fault_data_OR[index3-1,:] = seq_Fault_data_OR
  # Labeling Fault Outer Race output as 3

Output = np.array(Output)
Output = Output.reshape(4*M, 1)
print(Output.shape)
Data_matrix = np.concatenate((Array_Normal_data, Array_Fault_data_IR, Array_Fault_data_B, Array_Fault_data_OR), axis=0)
Data_matrix.shape
```

### Feature Extraction

```python
import scipy.stats as stats

def calculate_row_stats(Input_matrix):
  mean = np.mean(Input_matrix, axis=1)[:, np.newaxis]
  std = np.std(Input_matrix, axis=1)[:, np.newaxis]
  ptp = np.ptp(Input_matrix, axis=1)[:, np.newaxis]
  kurt = stats.kurtosis(Input_matrix, axis=1)[:, np.newaxis]
  skew = stats.skew(Input_matrix, axis=1)[:, np.newaxis]
  rms = np.sqrt(np.mean(Input_matrix**2, axis=1))[:, np.newaxis]
  # Calculate absolute minimum for each row
  abs_mean = np.amin(Input_matrix, axis=1)[:, np.newaxis]
  # Calculate impact factor (assuming difference between max and min)
  peak = np.amax(Input_matrix, axis=1)[:, np.newaxis]
  # Combine into a new matrix
  result_matrix = np.concatenate((mean, std, ptp, kurt, skew, rms, abs_mean, peak), axis=1)
  return result_matrix

stats_matrix = calculate_row_stats(Data_matrix)

# Print the resulting matrix (containing means and standard deviations)
print(stats_matrix.shape)
```

### Shuffle

```python
Class_shuffle, Target_shuffle = shuffle(stats_matrix, Output, random_state = 6)

Data_train, Data1, Target_train, Target1 = train_test_split(Class_shuffle, Target_shuffle, test_size = 0.3, random_state = 6)
Data_valid, Data_test, Target_valid, Target_test = train_test_split(Data1, Target1, test_size = 0.5, random_state = 6)

print(f"Data_train.shape: {Data_train.shape}, Target_train.shape: {Target_train.shape}")
print(f"Data_test.shape: {Data_test.shape}, Target_test.shape: {Target_test.shape}")
print(f"Data_valid.shape: {Data_valid.shape}, Target_valid.shape: {Target_valid.shape}")
```

```python
# Understand decision tree structure
n_nodes = clf.tree_.node_count
children_left = clf.tree_.children_left
children_right = clf.tree_.children_right
feature = clf.tree_.feature
threshold = clf.tree_.threshold
values = clf.tree_.value

node_depth = np.zeros(shape=n_nodes, dtype=np.int64)
is_leaves = np.zeros(shape=n_nodes, dtype=bool)
stack = [(0, 0)]  # start with the root node id (0) and its depth (0)
while len(stack) > 0:
    # `pop` ensures each node is only visited once
    node_id, depth = stack.pop()
    node_depth[node_id] = depth

    # If the left and right child of a node is not the same we have a split
    # node
    is_split_node = children_left[node_id] != children_right[node_id]
    # If a split node, append left and right children and depth to `stack`
    # so we can loop through them
    if is_split_node:
        stack.append((children_left[node_id], depth + 1))
        stack.append((children_right[node_id], depth + 1))
    else:
        is_leaves[node_id] = True

print(
    "Decision tree structure trained by original dataset \n"
    "The tree structure has {n} nodes and has "
    "the following tree structure:\n".format(n=n_nodes)
)
for i in range(n_nodes):
    if is_leaves[i]:
        print(
            "{space}node={node} is a leaf node with value={value}.".format(
                space=node_depth[i] * "\t", node=i, value=values[i]
            )
        )
    else:
        print(
            "{space}node={node} is a split node with value={value}: "
            "go to node {left} if X[:, {feature}] <= {threshold} "
            "else to node {right}.".format(
                space=node_depth[i] * "\t",
                node=i,
                left=children_left[i],
                feature=feature[i],
                threshold=threshold[i],
                right=children_right[i],
                value=values[i],
            )
        )

print('---------------------------------------------------')
```
